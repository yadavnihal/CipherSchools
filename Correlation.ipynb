# Correlation measures the strength and direction of a linear relationship between two variables the correlation cofficient(pearson's r) range varies from -1 to 1

```python
import numpy as np
import pandas as pd

# seed for productivity
np.random.seed(42)

data={
    "age":np.random.normal(30,10,100).astype(int),
    "annual income(k$)":np.random.normal(50,20,100).astype(int),
    "spending score":np.random.randint(1,100,100).astype(int),
    "years with company":np.random.normal(5,2,100).astype(int)
}

df=pd.DataFrame(data)
df

```

```python
from tabulate import tabulate
correlation_matrix=df.corr()
print(tabulate(correlation_matrix,headers="keys",tablefmt="grid",numalign="center",floatfmt=".2f"))
```
