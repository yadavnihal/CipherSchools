```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,5,9,16]
# creating a line plot
# plt.plot(x,y)
# plt.plot(x,y,linestyle="dotted")
# plt.plot(x,y,linestyle="dashed")
plt.plot(x,y,linestyle="dashdot",color="#ffe700",linewidth="3",marker="s",mec="r")#mfc for whole marker
# instead of writing linestyle we can also write ls as linestyle


plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo-Graph")
plt.show()

```

```python
import numpy as np
y1=np.array([1,2,3,4])
y2=np.array([1,4,9,16])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
# creating a scatter plot
x=[1,2,3,4]
y=[1,5,9,16]
plt.scatter(x,y,marker="s",color="green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo-Graph")
plt.show()
```

```python
# creating a bar graph
categories=["Monitor","Refrigirator","Laptop","Mobile"]
values=[6000,20000,40000,16000]
plt.bar(categories,values,color="purple",width=0.1)#use barh for horizontal view
plt.xlabel("Categories")
plt.ylabel("Price")
plt.title("Sales")
plt.show()
```

```python
#creating a histogram
data=[2,3,4,5,6,5,4,3]
plt.hist(data,color="orange",bins=4)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
# creating subplots with matplotlib
import matplotlib.pyplot as plt
x=[2,3,4,6,7]
y1=[1,5,9,16,20]
y2=[1,4,9,16,25]
fig ,axs=plt.subplots(2)
axs[0].plot(x,y1)
axs[0].set_title("First Plot")
axs[1].plot(x,y2)
axs[1].set_title("Second Plot")
plt.tight_layout()
plt.show()
```

```python
#Annotations in matplotlib
x=[2,3,4,6,7]
y=[1,5,9,16,20]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo-Graph")
plt.annotate("Peak",xy=(7,20),xytext=(5.5,17.5),arrowprops=dict(facecolor="black",shrink=0.05))
plt.show()
```

```python
# creating a simple line plot with seaborn
import seaborn as sns
x=[2,3,4,6,7]
y=[1,5,9,16,20]
sns.lineplot(x=x,y=y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("graph using seaborn")
plt.show()
```

```python
x=[2,3,4,6,7]
y=[1,5,9,16,20]
sns.scatterplot(x=x,y=y,color="red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("graph using seaborn")
plt.show()
```

```python

# creating pair plot using seaborn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
iris_data=sns.load_dataset("iris")

sns.pairplot(iris_data,hue="species")
plt.title("pair plot using seaborn")
plt.show()
```
