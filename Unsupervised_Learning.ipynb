## Unsupervised learning is a powerful machine learning technique that uncovers hidden patterns and insights from data without predefined label or target.

# K-Means Clustering
it aims to partition n observations into k clusters in such a way that each observation belongs to the cluster with the nearest mean,
the algorithm iteratively assigns the data points to a cluster and update the cluster until convergence.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# generating synthetic data
X, y = make_blobs(n_samples=300, centers=4,cluster_std=0.60,random_state=22)

#Training the data
model=KMeans(n_clusters=4)
y_pred=model.fit_predict(X)


# plotting the graph
plt.scatter(X[:,0],X[:,1],c=y_pred,cmap="rainbow")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=300,color="black",marker="X")
plt.title("K-mean clusturing")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Hierarchical Clustering
it is used to build a hierarchy of clusters. It can be divided into Agglomerative and Divisive clustering. The algorithm either starts with each element as a separate cluster and merges them to form a hierarchy (Agglomerative) or starts with a single cluster and divides it (Divisive).

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# generating synthetic data
X, y = make_blobs(n_samples=300, centers=4,cluster_std=0.60,random_state=42)

#training the data
model=AgglomerativeClustering(n_clusters=4)
y_pred=model.fit_predict(X)

# plotting the graph
plt.scatter(X[:,0],X[:,1],c=y_pred,cmap="rainbow")
plt.title("Hierarchical clusturing")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Principal Component Analysis (PCA)
Principal Component Analysis (PCA) is a dimensionality reduction technique. It transforms the data to a new coordinate system where the greatest variance by some projection of the data comes to lie on the first coordinate (called the first principal component), the second greatest variance on the second coordinate, and so on.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# loading the iris dataset
iris = load_iris()
X=iris.data
y=iris.target

# training the data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# plotting the graph
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,cmap="rainbow")
plt.title("PCA")
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")
plt.show()
```
